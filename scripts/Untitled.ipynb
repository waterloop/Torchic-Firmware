```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
df = pd.read_csv("./temp_data.csv")
display(df)
```

```python
%matplotlib widget

R = df["R"].to_numpy()
T = df["T"].to_numpy()

coeffs = np.polyfit(R, T, 4)
coeffs = np.flip(coeffs)
display(coeffs)

def reg_eq(r):
    ret = 0
    for i, c in enumerate(coeffs):
        ret += c*(r**i)
    return ret

plt.close()
plt.scatter(R, T)
plt.plot(R, [reg_eq(r) for r in R])
plt.show()
```
